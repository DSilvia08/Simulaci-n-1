```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def estimación(N):
 l=[]
 éxitos=0
 for i in range(N):
  u_1=rnd.random()
  u_2=rnd.random()
  if u_2<=np.sqrt(1-u_1**2):
   éxitos+=1
  prob=éxitos/(i+1)
  l.append(4*prob)
 return l
```
