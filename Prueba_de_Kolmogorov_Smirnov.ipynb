```python
#Tomando la lista del ejercicio 3
L=[0.8797, 0.3884, 0.6289, 0.8750, 0.5999, 0.8589, 0.9996, 0.2415, 0.3808, 0.9606,
0.9848, 0.3469, 0.7977, 0.5844, 0.8147, 0.6431, 0.7387, 0.5613, 0.0318, 0.7401,
0.4557, 0.1592, 0.8536, 0.8846, 0.3410, 0.1492, 0.8681, 0.5291, 0.3188, 0.5992,
0.9170, 0.2204, 0.5991, 0.5461, 0.5739, 0.3254, 0.0856, 0.2258, 0.4603, 0.5027,
0.8376, 0.6235, 0.3681, 0.2088, 0.1525, 0.2006, 0.4720, 0.4272, 0.6360, 0.0954]

#1. Ordenar de menor a mayor los números aleatorios
L.sort()
print(L)
j=len(L)

#2.
#Calcular D+.    max 1<i<n |i/n-x_i|
DM=[]
i=0
for n in L:
  S=abs(((i/j)-n))
  DM.append(round(S,4))
  i=i+1

Y=max(DM)
print("D+ = ", Y)

#D^-.    max 1<i<n  |x_i- (i-n)/n|
k=0
Dm=[]
for n in L:
  S=abs((n-((k-1)/j)))
  Dm.append(round(S,4))
  k=k+1

Z=max(Dm)
print("D- = ",Z)

#Tabla
!pip install prettytable
from prettytable import PrettyTable
myTable = PrettyTable(["i","X_i", "D+", "D-"])

h=1
for x in range (j):
  myTable.add_row([h,L[x],DM[x],Dm[x]])
  h=h+1
print(myTable)

#D= max {D+,D-}
print("D= max {D+,D-} : ", max(Y,Z))
#3. Comparamos con el valor crítico d_on


#se acepta di D<d_on
```
